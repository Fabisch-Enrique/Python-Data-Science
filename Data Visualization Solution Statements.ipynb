```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100)
```

```python
x
```

```python
y = x*2
```

```python
y
```

```python
z = x**2
```

```python
z
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

fig
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax.set_title('ZOOM')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

# SUBPLOTS

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x,y,color='blue', lw=3, ls='--')
axes[1].plot(x,y, color='red', lw=3, ls='-')

fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

axes[0].plot(x,y, color='red', lw=5, ls='--')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(x,z, color='purple', lw=3)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
```
